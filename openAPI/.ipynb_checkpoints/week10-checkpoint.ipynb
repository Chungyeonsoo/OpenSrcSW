```python
print("encryption : 0")
print("decryption : 1")
x = input()

if x == '0':
    code = []
    
    print("input to encrypt :")
    encrypt = input()
    
    se = encrypt.split()

    for a in se:
        string = ""
        
        for b in a:
            set = ord(b) + 1
            b = chr(set)
            string += b
            
        code.append(string)
        
    code.reverse()        
    print(code)
    #암호화
    
elif x == '1':
    decode = []
        
    print("input to decrypt :")
    decrypt = input()
    
    sd = decrypt.split()

    for a in sd:
        string = ""
        
        for b in a:
            set = ord(b) - 1
            b = chr(set)
            string += b
            
        decode.append(string)
        
    decode.reverse()        
    print(decode)
    
    #복호화
else:
    print("wrong input")
```

```python
phonebook = {}
dict = {}

while True:
    print("input data : 0")
    print("search data : 1")
    print("anything else to break")

    function = input()

    if function == '0':
        print("press 'q' to quit.")
        print("input name & phonenumber :")
    
        while True:
            x = input()
            if x == 'q':
                print(" ")
                break
            y = x.split()
            phonebook[y[0]] = y[1] 
        

    elif function == '1':
        print("press 'q' to quit.")
        print("1 - search name :")
        print("2 - search phone number :")
        while True:
            x = input()
            if x == 'q':
                print(" ")
                break
            elif x == '1':
                print("name input:")
                name = input()
                print("phone number of", name, " = ", phonebook[name])
            elif x == '2':
                print("phone number input:")
                phonenum = input()
                for k, v in phonebook.items():
                    if v == phonenum:
                        print(k)
                

    else:
        print("wrong input. end program")
        break
```
