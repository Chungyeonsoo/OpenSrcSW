```python
print("encryption : 0")
print("decryption : 1")
x = input()

if x == '0':
    code = []
    
    print("input to encrypt :")
    encrypt = input()
    
    se = encrypt.split()

    for a in se:
        string = ""
        
        for b in a:
            set = ord(b) + 1
            b = chr(set)
            string += b
            
        code.append(string)
        
    code.reverse()        
    print(code)
    #암호화
    
elif x == '1':
    decode = []
        
    print("input to decrypt :")
    decrypt = input()
    
    sd = decrypt.split()

    for a in sd:
        string = ""
        
        for b in a:
            set = ord(b) - 1
            b = chr(set)
            string += b
            
        decode.append(string)
        
    decode.reverse()        
    print(decode)
    
    #복호화
else:
    print("wrong input")
```

```python
phonebook = {}

while True:
    print("input data : 0")
    print("search data : 1")
    print("anything else to break")

    function = input()

    if function == '0':
        print("press 'q' to quit.")
        print("input name & phonenumber :")
    
        while True:
            x = input()
            if x == 'q':
                print(" ")
                break
            y = x.split()
            phonebook[y[0]] = y[1] 
        

    elif function == '1':
        print("press 'q' to quit.")
        print("input name :")
        while True:
            name = input()
            if name == 'q':
                print(" ")
                break
            print("phone number of", name, " = ", phonebook[name])

    else:
        print("wrong input. end program")
        break
```
