```python
# csv 파일 변환 문제로 인해 split '-' --> '.'

import csv
import matplotlib.pyplot as plt

f = open('/Users/chungyeons/Downloads/seoul2.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)

min_result = []
max_result = []
year = []

for row in data:
    if row[-2] != '' and row[-1] != '':
        if row[0].split('.')[1] == '8' and row[0].split('.')[2] == '31':
            max_result.append(float(row[-1]))
            min_result.append(float(row[-2]))
            year.append(row[0].split('.')[0])

plt.rc('font', family='AppleGothic')

plt.title("생일 기온 변화")
plt.plot(year[-10:], max_result[-10:], label ="최고 기온")
plt.plot(year[-10:], min_result[-10:], label ="최저 기온")
plt.legend()

plt.show()
        
```

```python
import csv
import matplotlib.pyplot as plt
f = open('/Users/chungyeons/Downloads/seoul2.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)

jan_result = []
aug_result = []

for row in data:
    if row[-2] != '' and row[-1] != '':
            if row[0].split('.')[1] == '1':
                jan_result.append(float(row[-1]))
            elif row[0].split('.')[1] == '8':
                aug_result.append(float(row[-1]))
                
        

plt.hist(jan_result, bins =100, color="blue", label='Jan')
plt.hist(aug_result, bins =100, color="red", label='Aug')
plt.legend()

plt.show()
```
