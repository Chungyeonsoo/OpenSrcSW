```python
ALL = []
recipient_info = ''
recipient_arry = []
pre_rec_arr = []
product_name = ''
product_opt = ''
product_cnt = ''
product = []
delivery_msg = ''
ALL = []
product_set =''
j = 0
pre_set = []
pd_cv = ''

for i in range (0,length):
    pre_set = []
    recipient_name = df_from_excel['수취인명'][i]
    recipient_numb = df_from_excel['수취인연락처1'][i]
    recipient_addr = df_from_excel['배송지'][i]
    product_name = df_from_excel['상품명'][i]
    if product_name == '링꼬테이블 따뜻한 와인 건조과일 뱅쇼키트 뱅쇼만들기 45g':
        product_name = '뱅쇼키트'
    
    product_opt = df_from_excel['옵션정보'][i]
    
    if product_opt == '향신료: 로즈마리' or product_opt == '향신료: 시나몬스틱 3개' or product_opt == '':
        product_opt = ' '
    delivery_msg = df_from_excel['배송메세지'][i]
    product_cnt = df_from_excel['수량'][i]
    pre_set.append([recipient_name, recipient_numb, recipient_addr, delivery_msg])

    
    # 첫번째
    if i == 0:
        # 첫번째 pra에 저장
        # productset에 상품명+옵션 저장
        #pre-rec 주소 저장
        #pre_rec_arr = recipient_arry
        
        recipient_arry = pre_set
        product_set = str(product_name) + str(product_opt) + str(' x') + str(product_cnt) + str(' / ')

    # 중복 고객일시    
    elif recipient_arry[0][2] == recipient_addr:
        # 전 productset을 product에 저장, productset 초기화
        product.append(product_set)
        product_set = str(product_name) + str(product_opt) + str(' x') + str(product_cnt) + str(' / ')
        
        # 중복 고객 마지막
        if i == length -1:
            product_set = str(product_name) + str(product_opt) + str(' x') + str(product_cnt) + str(' / ')
            product.append(product_set)
            for j in range(0, len(product)):
                
                pd_cv = pd_cv + product[j]
            pd_cv = pd_cv.replace('nan',"") 
            ALL.append([recipient_arry[0][0], recipient_arry[0][1], recipient_arry[0][2], pd_cv, recipient_arry[0][3]])
            pd_cv = ''

    # 중복 x 고객     
    elif recipient_arry[0][2] != recipient_addr and i > 0:
        product.append(product_set)
        for j in range(0, len(product)):
            pd_cv = pd_cv + product[j]
        pd_cv = pd_cv.replace('nan',"") 
        ALL.append([recipient_arry[0][0], recipient_arry[0][1], recipient_arry[0][2], pd_cv, recipient_arry[0][3]])
        pd_cv = ''
        # 초기화
        recipient_arry = pre_set
        product = []
        pd_dv = ''
        
        # 마지막
        if i == length -1:
            product_set = str(product_name) + str(product_opt) + str(' x') + str(product_cnt) + str(' / ')
            product.append(product_set)
            
            ###

            for j in range(0, len(product)):
                pd_cv = pd_cv + product[j]
            pd_cv = pd_cv.replace('nan',"") 
            ###
            ALL.append([recipient_arry[0][0], recipient_arry[0][1], recipient_arry[0][2], pd_cv, recipient_arry[0][3]])
            pd_cv = ''
        else :
            product_set = str(product_name) + str(product_opt) + str(' x') + str(product_cnt) + str(' / ')
```
